```python
import random

raw_data = [random.randint(1, 100) for _ in range(10)]
print(f"Week 1 (The Data): {raw_data}")

def bubble_sort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
    return data

sorted_data = bubble_sort(raw_data.copy())
print(f"Week 2 (Sorted):    {sorted_data}")


def binary_search(data, target):
    low = 0
    high = len(data) - 1
    while low <= high:
        mid = (low + high) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return - 1

target = sorted_data[2]
result = binary_search(sorted_data, target)
print(f"Week 3 (Search):    Found {target} at index {result}")

class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key
    
def insert(root, key):
        if root is None:
            return Node(key)
        else:
            if root.val < key:
                root.right = insert(root.right, key)
            else:
                root.left = insert(root.left, key)
            return root
        
root = Node(sorted_data[0])
for i in range(1, len(sorted_data)):
    insert(root, sorted_data[i])
    
print("Week 4 (Tree):       Binary Tree built successfully!")
            
```
