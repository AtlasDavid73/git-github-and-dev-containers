```python
def linear_search_for(target, dataset):
    for index in range(len(dataset)):
        if dataset[index] == target:
            return index
    return -1

data = [4, 7, 1, 9, 3]
print(linear_search_for(9, data))
print(linear_search_for(5, data))
```

```python
def linear_search_while(target, dataset):
    idx = 0
    while idx < len(dataset):
        if dataset[idx] == target:
            return idx
        idx += 1
    return -1

data = [10, 20, 30, 40]
print(linear_search_while(30,data))
print(linear_search_while(5, data))
```

```python
def linear_search_exit(target, dataset):
    for idx, value in enumerate(dataset):
        if value == target:
            return idx
    return -1

data = [2, 4, 6, 8, 10]
print(linear_search_exit(2, data))
print(linear_search_exit(10, data))
```

```python
def linear_search_sorted(target, dataset):
    for idx, value in enumerate(dataset):
        if value == target:
            return idx
        if value > target:
            break
    return -1

data = [1, 3, 5, 7, 9, 11]
print(linear_search_sorted(7, data))
print(linear_search_sorted(8, data))       
```
