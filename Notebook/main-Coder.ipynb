```python
def add(a, b):
    """Simple addition function"""
    return a + b

def test_add():
    """Test the add function"""
    assert add(2, 3) == 5
    assert add(-1, 1) == 0
    assert add(0, 0) == 0
    print("All tests passed!")

# Run the test
test_add()
```

```python
# Assign two numbers to variables
num1 = 5
num2 = 10

# Add the numbers using the '+' operator
sum = num1 + num2

# Print the result
print("The sum of", num1, "and", num2, "is:", sum)
```
